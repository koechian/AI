## Clustering

Use k as either 3, 4 OR 5. (Only pick one of the three)

```python
# Use k - Means Clustering to cluster our data and assign each row to a cluster (assing labels to each row). 
# Be carefull with the arguments you pass to the function, make sure you pass the correct
from sklearn.cluster import KMeans

k = 5
kmean=KMeans(n_clusters=k)
kmean.fit(Clustering_df)
kmean.cluster_centers_

# create a new column in our data and assign the cluster labels to each row: this has been partially done for you

Clustering_df['Clusters']=kmean.labels_
Clustering_df.head(20)

```

```python
# train a kNN model to predict the cluster of each row in our test data
# Be careful with the arguments you pass to the function, make sure you pass the correct
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# Create an instance of the KNeighborsClassifier class. Set the number of neighbors to 3
kncc = KNeighborsClassifier(n_neighbors=3)

# Specify the columns to use for training i.e. Features (X) and the target column (y). Remeber the tagret is the cluster column from the clustering step

# Splitting the data into train and test models 
# splits 70% of the data for trainning and 30% for testing
train, test = train_test_split(Clustering_df,test_size = 0.3)

X = train.drop('Clusters',axis= 1)
y = train['Clusters']

# Train / fit Features and Taregts to the KNeighborsClassifier model
kncc.fit(X,y)
predictions = kncc.predict(test.drop('Clusters',axis = 1))

print(accuracy_score((test['Clusters']),predictions))
```
